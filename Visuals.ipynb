```python
import matplotlib.pyplot as plt

# Define the data
races = ['African American', 'Caucasian', 'Hispanic', 'Other', 'Asian', 'Native American']
real_error_rates = [0.26, 0.4, 0.46, 0.36, 0.75, 0.5]
synthetic_error_rates = [0.15, 0.16, 0.13, 0.16, 0.13, 0]

# Plot the graph
plt.figure()
plt.title('Error Rate - For Each Race FNR')
plt.ylabel('Error Rate')
plt.xlabel('Data Source')

# Plot the lines connecting the "real" and "synthetic" error rates for each race
for i in range(len(races)):
    race = races[i]
    real_error = real_error_rates[i]
    synthetic_error = synthetic_error_rates[i]

    plt.plot([1, 2], [real_error, synthetic_error], marker='o', label=race)

plt.xticks([1, 2], ['Real', 'Synthetic'])
plt.ylim(0, 1.1)  # Adjust the y-axis limit to end at 1
plt.legend()
plt.show()
```
